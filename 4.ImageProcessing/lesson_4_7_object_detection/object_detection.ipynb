```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("input/wolf.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap="gray")
plt.show()
```

```python
_, image_thresh = cv2.threshold(image, 115, 255, cv2.THRESH_BINARY)
plt.imshow(image_thresh, cmap="gray")
plt.show()
```

```python
contours, _ = cv2.findContours(image_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print(contours)
```

```python
result = image.copy()

for contour in contours:
    if cv2.contourArea(contour) > 1000:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(result, (x, y), (x+w, y+h), (255, 255, 255), 3)

plt.imshow(result, cmap="gray")
plt.show()
```

```python
image_puma = cv2.imread("input/puma.jpg", cv2.IMREAD_GRAYSCALE)

plt.imshow(image_puma, cmap="gray")
```

```python
image_puma = 255 - image_puma

plt.imshow(image_puma, cmap="gray")
plt.show()
```

```python
_, image_puma_thresh = cv2.threshold(image_puma, 115, 255, cv2.THRESH_BINARY)
```

```python
contours, _ = cv2.findContours(image_puma_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
result = image_puma.copy()
for contour in contours:
    if cv2.contourArea(contour) > 10:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(result, (x, y), (x+w, y+h), (127, 127, 127), 3)

plt.imshow(result, cmap="gray")
```
