```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')
```

```python
face_a_1 = 1.4
face_a_2 = 1.5
face_b_1 = 4.1
face_b_2 = 4.4
face_b_3 = 4.6
face_c_1 = 6.0
face_c_2 = 6.2

plt.figure(figsize=(16, 16))
plt.arrow(0, 0, face_a_1, 0, head_width=0.1, head_length=0.1, fc='red', ec='red', width=0.04)
plt.arrow(0, 0, face_a_2, 0, head_width=0.1, head_length=0.1, fc='red', ec='red', width=0.04)
plt.arrow(0, 0, face_b_1, 0, head_width=0.1, head_length=0.1, fc='blue', ec='blue', width=0.02)
plt.arrow(0, 0, face_b_2, 0, head_width=0.1, head_length=0.1, fc='blue', ec='blue', width=0.02)
plt.arrow(0, 0, face_b_3, 0, head_width=0.1, head_length=0.1, fc='blue', ec='blue', width=0.02)
plt.arrow(0, 0, face_c_1, 0, head_width=0.1, head_length=0.1, fc='green', ec='green', width=0.01)
plt.arrow(0, 0, face_c_2, 0, head_width=0.1, head_length=0.1, fc='green', ec='green', width=0.01)


plt.xlim(-1, 7)
plt.ylim(-1, 1)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
