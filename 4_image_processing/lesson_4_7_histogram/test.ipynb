```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("input/a1.tif", cv2.IMREAD_GRAYSCALE)
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()

image = cv2.imread("input/a2.tif", cv2.IMREAD_GRAYSCALE)
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()

image = cv2.imread("input/a3.tif", cv2.IMREAD_GRAYSCALE)
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
image_low_contrast = cv2.imread("input/fingerprint.png", cv2.IMREAD_GRAYSCALE)
hist = cv2.calcHist([image_low_contrast], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
image_high_contrast = cv2.equalizeHist(image_low_contrast)
cv2.imwrite("output/result11.jpg", image_high_contrast)
hist = cv2.calcHist([image_high_contrast], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```
